```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data for world
df_world = pd.read_csv("EV_WORLD.csv")
df_world = df_world[df_world['powertrain'] == 'BEV']
df_world = df_world.drop(['category', 'parameter', 'mode', 'unit', 'region', 'powertrain'], axis=1)

# Load the data for India
df_india = pd.read_csv("EV_INDIA_SALES.csv")
df_india = df_india[df_india['powertrain'] == 'BEV']
df_india = df_india.drop(['category', 'parameter', 'mode', 'unit', 'region', 'powertrain'], axis=1)

# Load the data for Europe
df_europe = pd.read_csv("EV_Europe.csv")
df_europe = df_europe[df_europe['powertrain'] == 'BEV']
df_europe = df_europe.drop(['category', 'parameter', 'mode', 'unit', 'region', 'powertrain'], axis=1)

# Plot the data
plt.plot(df_world['year'], df_world['value'] / 1000, color='blue', linewidth=2, label='World')
plt.scatter(df_world['year'], df_world['value'] / 1000, color='blue', marker='o', s=50, zorder=10)

plt.plot(df_india['year'], df_india['value'] / 1000, color='red', linewidth=2, label='India')
plt.scatter(df_india['year'], df_india['value'] / 1000, color='red', marker='o', s=50, zorder=10)

plt.plot(df_europe['year'], df_europe['value'] / 1000, color='green', linewidth=2, label='Europe')
plt.scatter(df_europe['year'], df_europe['value'] / 1000, color='green', marker='o', s=50, zorder=10)

# Customize the plot
plt.xlabel('Year', fontsize=16)
plt.ylabel('Value in thousands', fontsize=16)
plt.title('Value vs Year of BEV sales', fontsize=16)
plt.legend()
plt.rcParams["legend.fontsize"] = 12

# Increase font size of axis labels
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)


# Display the plot
plt.show()
```
