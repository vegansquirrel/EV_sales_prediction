```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data for world
df_world = pd.read_csv("EV_WORLD.csv")
df_world_bev = df_world[df_world['powertrain'] == 'BEV']
df_world_phev = df_world[df_world['powertrain'] == 'PHEV']
df_world_bev = df_world_bev.drop(['category', 'parameter', 'mode', 'unit', 'region', 'powertrain'], axis=1)
df_world_phev = df_world_phev.drop(['category', 'parameter', 'mode', 'unit', 'region', 'powertrain'], axis=1)

# Load the data for India
df_india = pd.read_csv("EV_INDIA_SALES.csv")
df_india_bev = df_india[df_india['powertrain'] == 'BEV']
df_india_phev = df_india[df_india['powertrain'] == 'PHEV']
df_india_bev = df_india_bev.drop(['category', 'parameter', 'mode', 'unit', 'region', 'powertrain'], axis=1)
df_india_phev = df_india_phev.drop(['category', 'parameter', 'mode', 'unit', 'region', 'powertrain'], axis=1)

# Load the data for Europe
df_europe = pd.read_csv("EV_Europe.csv")
df_europe_bev = df_europe[df_europe['powertrain'] == 'BEV']
df_europe_phev = df_europe[df_europe['powertrain'] == 'PHEV']
df_europe_bev = df_europe_bev.drop(['category', 'parameter', 'mode', 'unit', 'region', 'powertrain'], axis=1)
df_europe_phev = df_europe_phev.drop(['category', 'parameter', 'mode', 'unit', 'region', 'powertrain'], axis=1)

# Plot the data for world
plt.figure(figsize=(10, 6))
plt.plot(df_world_bev['year'], df_world_bev['value'] / 1000, color='blue', linewidth=2, label='BEV')
plt.scatter(df_world_bev['year'], df_world_bev['value'] / 1000, color='blue', marker='o', s=75, zorder=10)

plt.plot(df_world_phev['year'], df_world_phev['value'] / 1000, color='red', linewidth=2, label='PHEV')
plt.scatter(df_world_phev['year'], df_world_phev['value'] / 1000, color='red', marker='o', s=75, zorder=10)

plt.xlabel('Year', fontsize=20)
plt.ylabel('Value in thousands', fontsize=20)
plt.title('BEV and PHEV Sales in the World', fontsize=20)
# Increase font size of axis labels
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

plt.legend()
plt.rcParams["legend.fontsize"] = 20
plt.show()

# Plot the data for India
plt.figure(figsize=(10, 6))
plt.plot(df_india_bev['year'], df_india_bev['value'] / 1000, color='blue', linewidth=2, label='BEV')
plt.scatter(df_india_bev['year'], df_india_bev['value'] / 1000, color='blue', marker='o', s=75, zorder=10)

plt.plot(df_india_phev['year'], df_india_phev['value'] / 1000, color='red', linewidth=2, label='PHEV')
plt.scatter(df_india_phev['year'], df_india_phev['value'] / 1000, color='red', marker='o', s=75, zorder=10)

plt.xlabel('Year', fontsize=20)
plt.ylabel('Value in thousands', fontsize=20)
plt.title('BEV and PHEV Sales in India', fontsize=20)
# Increase font size of axis labels
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

plt.legend()
plt.rcParams["legend.fontsize"] = 20
plt.show()

# Plot the data for Europe
plt.figure(figsize=(10, 6))
plt.plot(df_europe_bev['year'], df_europe_bev['value'] / 1000, color='blue', linewidth=2, label='BEV')
plt.scatter(df_europe_bev['year'], df_europe_bev['value'] / 1000, color='blue', marker='o', s=75, zorder=10)

plt.plot(df_europe_phev['year'], df_europe_phev['value'] / 1000, color='red', linewidth=2, label='PHEV')
plt.scatter(df_europe_phev['year'], df_europe_phev['value'] / 1000, color='red', marker='o', s=75, zorder=10)

plt.xlabel('Year', fontsize=20)
plt.ylabel('Value in thousands', fontsize=20)
plt.title('BEV and PHEV Sales in Europe', fontsize=20)
# Increase font size of axis labels
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

plt.legend()
plt.rcParams["legend.fontsize"] = 20
plt.show()
```
