```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

x=pd.read_csv("australia_EV.csv")
x = x.drop(x.columns[0], axis=1)
x
```

made seperate dataset for usable data

```python
years = x[['year']]
values=x[['value']]

# Create and train the linear regression model
model = LinearRegression()
model.fit(years, values)


# Predict the scaled value for future years
future_years = [[2023], [2024], [2025]]  # Example future years
scaled_predictions = model.predict(pd.DataFrame(future_years, columns=['year']))


# Create a dataframe to display the predictions
predictions_df = pd.DataFrame({'year': [year[0] for year in future_years], 'predicted_value': scaled_predictions.flatten()})
print(predictions_df)
```

```python
import matplotlib.pyplot as plt

plt.xlabel='year'
plt.ylabel='value'
plt.scatter(predictions_df.year,predictions_df.predicted_value,color='red',marker='+')
plt.plot(predictions_df.year,predictions_df.predicted_value,color='blue')
plt.show()
```
